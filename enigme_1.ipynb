```python
from qiskit import QuantumCircuit, Aer, transpile
from qiskit.visualization import plot_histogram
```

```python
quantum_circuit = QuantumCircuit(3)
quantum_circuit.draw("mpl")
```

```python
quantum_circuit.h(0)
quantum_circuit.draw("mpl")
```

```python
quantum_circuit.cx(0,1)
quantum_circuit.draw("mpl")
```

```python
quantum_circuit.h(2)
quantum_circuit.draw("mpl")
```

```python
quantum_circuit.cx(2,1)
quantum_circuit.x(2)
quantum_circuit.draw("mpl")
```

```python
quantum_circuit.cx(2,0)
quantum_circuit.x(2)
quantum_circuit.draw("mpl")
```

```python
quantum_circuit.swap(0,1)
quantum_circuit.draw("mpl")
```

```python
quantum_circuit.x(0)
quantum_circuit.x(1)
quantum_circuit.draw("mpl")
```

```python
quantum_circuit.cx(2,1)
quantum_circuit.x(2)
quantum_circuit.draw("mpl")
```

```python
quantum_circuit.cx(2,0)
quantum_circuit.x(2)
quantum_circuit.draw("mpl")
```

```python
backend = Aer.get_backend("statevector_simulator")
```

```python
run = backend.run(quantum_circuit)
result = run.result()
```

```python
output_state = result.get_statevector(quantum_circuit, decimals=3)
print(output_state)
```

```python
measures = QuantumCircuit(3, 3)
measures.measure([0, 1, 2], [0, 1, 2])

backend = Aer.get_backend('qasm_simulator')
q_circ = quantum_circuit.compose(measures)
result = backend.run(transpile(q_circ, backend), shots=10000).result()
counts = result.get_counts()
```

```python
plot_histogram(counts)
```
